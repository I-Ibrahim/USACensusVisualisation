```python
Population=pd.concat([Pop1980, Pop1990,Pop2000,Pop2010], axis=1)
Population.loc['United States']=pd.Series({'Population 1980':'226545805', 'Population 1990':'248709873',
                                      'Population 2000':'281421906', 'Population 2010':'311536594' })
PopGDP=pd.concat([Population, GDP], axis=1)
PopGDP=PopGDP.dropna(axis=0, how='any')
PopGDP.index.name='State'
```

```python
PopGDP = PopGDP.astype(float).fillna(0.0)
```

```python
PopGDP['GDP per cap 1980']=PopGDP['GDP 1980']/PopGDP['Population 1980']
PopGDP['GDP per cap 1990']=PopGDP['GDP 1990']/PopGDP['Population 1990']
PopGDP['GDP per cap 2000']=PopGDP['GDP 2000']/PopGDP['Population 2000']
PopGDP['GDP per cap 2010']=PopGDP['GDP 2010']/PopGDP['Population 2010']

PopGDP['GDP per cap 1980 normalised']=(PopGDP['GDP per cap 1980']/PopGDP['GDP per cap 1980'])*100
PopGDP['GDP per cap 1990 normalised']=(PopGDP['GDP per cap 1990']/PopGDP['GDP per cap 1980'])*100
PopGDP['GDP per cap 2000 normalised']=(PopGDP['GDP per cap 2000']/PopGDP['GDP per cap 1980'])*100
PopGDP['GDP per cap 2010 normalised']=(PopGDP['GDP per cap 2010']/PopGDP['GDP per cap 1980'])*100


PopGDP['GDP 1980 normalised']=((PopGDP['GDP 1980']/PopGDP['Population 1980'])/PopGDP['GDP per cap 1980'])*100
PopGDP['GDP 1990 normalised']=((PopGDP['GDP 1990']/PopGDP['Population 1980'])/PopGDP['GDP per cap 1980'])*100
PopGDP['GDP 2000 normalised']=((PopGDP['GDP 2000']/PopGDP['Population 1980'])/PopGDP['GDP per cap 1980'])*100
PopGDP['GDP 2010 normalised']=((PopGDP['GDP 2010']/PopGDP['Population 1980'])/PopGDP['GDP per cap 1980'])*100
```

# Gain vs Loss

```python
Gain2010=Census2010.ix[:,['medianEarnings']]
Gain1980=Census1990.ix[:,['medianEarnings']]
Gain2000=Census2000.ix[:,['medianEarnings']]



Gain2010=Gain2010.rename(columns={'medianEarnings':'2010 Median'})
Gain1980=Gain1980.rename(columns={'medianEarnings':'1980 Median'})
Gain2000=Gain2000.rename(columns={'medianEarnings':'2000 Median'})




```

```python
Gain=pd.concat([Gain2010, Gain1980, Gain2000], axis=1)
Gain.head()
```

```python
Gain['Diff']=Gain['2010 Median']-Gain['1980 Median']
Gain['Diff 2']=Gain['2010 Median']-Gain['2000 Median']
Gain.head()
```

```python
Gain.loc[Gain['Diff']>0, 'Gain/Loss']='Gain'
Gain.loc[Gain['Diff']<0, 'Gain/Loss']='Loss'
```
